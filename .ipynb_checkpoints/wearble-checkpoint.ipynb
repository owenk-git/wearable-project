```python

import pandas as pd
import numpy as np


# file_path = pwd_ + "/ref_data/lab_data/t2_treadmill_walking_001-000_00B4D7FF.txt"
file_path = pwd_ + "/ref_data/lab_data/t2_treadmill_walking_001-000_00B4D7D6.txt"

# We'll define 14 column names (ignoring Quat_q3 since it’s missing in the data)
col_names = [
    "PacketCounter",
    "SampleTimeFine",
    "Acc_X",
    "Acc_Y",
    "Acc_Z",
    "Gyr_X",
    "Gyr_Y",
    "Gyr_Z",
    "Mag_X",
    "Mag_Y",
    "Mag_Z",
    "Quat_q0",
    "Quat_q1",
    "Quat_q2"
]

df = pd.read_csv(
    file_path,
    sep=r"\s+",         # handles multiple spaces/tabs
    engine="python",
    comment='//',       # skip lines that start with "//"
    names=col_names,    # use our custom 14-column list
    header=None,        # don’t treat any row as header
    skiprows=5          # skip the first 5 lines (including the 15-column header)
)

# print("Columns:", df.columns)
# print(df.head())

# Now you can slice just Acc_X, Acc_Y, Acc_Z
acc_data = df[["Acc_X", "Acc_Y", "Acc_Z"]]
acc_array = acc_data.to_numpy()
# or acc_array = acc_data.values

print(acc_array)
print(acc_array.shape)

angle_left = acc_array
```

```python
import numpy as np
from scipy.interpolate import interp1d

# Original sampling rate and desired upsampling factor
original_rate = 40  # Hz
target_rate = 200   # Hz
upsample_factor = target_rate // original_rate

# Original time vector
t_left = np.arange(angle_left.shape[0]) / original_rate

# New time vector for upsampled data
t_left_upsampled = np.linspace(t_left[0], t_left[-1], angle_left.shape[0] * upsample_factor)

# Interpolate each axis separately
angle_left_upsampled = np.zeros((len(t_left_upsampled), angle_left.shape[1]))
for i in range(angle_left.shape[1]):
    interpolator = interp1d(t_left, angle_left[:, i], kind='linear')
    angle_left_upsampled[:, i] = interpolator(t_left_upsampled)
```

```python

# Update Rate: 40.0Hz,  200 Hz

import numpy as np
import matplotlib.pyplot as plt

t_ori = np.arange(angle_ori.shape[0])
t_left = np.arange(angle_left_upsampled.shape[0])

fig, axs = plt.subplots(3, 1, figsize=(12, 8), sharex=True)
titles = ['X ACC', 'Y ACC', 'Z ACC']

for i in range(3):
    axs[i].plot(t_ori, angle_ori[:, i], label='calgary_data/lshake', alpha=0.7)
    axs[i].plot(t_left, angle_left_upsampled[:, i], label='lab_data/LeftShak', alpha=0.7)
#     axs[i].set_ylabel('Accelerator')
    axs[i].set_title(titles[i])
    axs[i].legend()

plt.show()
```
